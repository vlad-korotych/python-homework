Загрузить файл wines_data.csv

```python
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
wines = pd.read_csv('wine_data.csv')
wines.head()
```

```python
wines.info()
```

```python
wines.describe()
```

Рассчитать и визуализировать корреляции для переменных данного датасета

```python
corr = wines.corr()
corr
```

```python
sns.heatmap(corr, annot=True,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

Визуализировать взаимосвязи между 3-4 переменными

```python
sns.pairplot(wines, vars=['Flavanoids', 'Total phenols'], kind='scatter')
```

```python
sns.pairplot(wines, vars=['Nonflavanoid phenols', 'OD280/OD315 of diluted wines'], kind='scatter')
```

```python
sns.pairplot(wines, vars=['Alcohol', 'Hue'], kind='scatter')
```

```python
sns.pairplot(wines, vars=['Flavanoids', 'Class'], kind='scatter')
```

Рассчитать t-test статистики для 3-4 переменных

```python
stats.ttest_ind(wines['Flavanoids'], wines['Total phenols'])
```

```python
stats.ttest_ind(wines['Nonflavanoid phenols'], wines['OD280/OD315 of diluted wines'])
```

```python
stats.ttest_ind(wines['Flavanoids'], wines['Hue'])
```

```python
stats.ttest_ind(wines['Alcohol'], wines['Class'])
```
